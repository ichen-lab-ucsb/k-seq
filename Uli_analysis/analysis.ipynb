```python
import matplotlib.pyplot as plt

timePoints = [1.5, 5, 16.5, 55, 180]
colorList = plt.rcParams['axes.prop_cycle'].by_key()['color']
markerList = ['o', 's', '^', '*', 'D', 'x', '>', 'p', '1', 'X']

fig,ax = plt.subplots(1,2,figsize=[11,5])
for seqIx,seq in enumerate(expData):
    ax[0].plot(timePoints, seq['ligPct'], '%s-'%markerList[seqIx])
    ax[1].plot(timePoints, seq['kSeqRecRatio'], '%s-'%markerList[seqIx], label=seq['id'])

ax[0].set_xlabel('Reaction time (min)', fontsize=14)
ax[0].set_ylabel('Ligation ratio', fontsize=14)
ax[0].text(-5, 0.28, 'Gel shift experiment', fontsize=14)
ax[1].set_xlabel('Reaction time (min)', fontsize=14)
ax[1].set_ylabel('Recovery ratio', fontsize=14)
ax[1].text(-5, 0.28, 'k-Seq', fontsize=14)
ax[1].legend(loc=[1,0.2],frameon=False)
plt.tight_layout()
plt.show()
```

```python
def get_fitting_line(x,y):
    from scipy.optimize import curve_fit
    def func(x,a,b):
        return a*x+b
    try:
        params, pcov = curve_fit(func, xdata=x, ydata=y, method='trf')
    except RuntimeError:
        params = [np.nan, np.nan]
    return params


colorList = plt.rcParams['axes.prop_cycle'].by_key()['color']
markerList = ['o', 's', '^', '*', 'D', 'x', '>', 'p', '1', 'X']

plt.figure(figsize=[8,8])

for seqIx,seq in enumerate(expData):
    plt.scatter(seq['ligPct'], seq['kSeqRecRatio'], marker=markerList[seqIx], facecolor=colorList[seqIx], label=seq['id'])
    params = get_fitting_line(seq['ligPct'], seq['kSeqRecRatio'])
    plt.plot([0, 0.3], [params[1], 0.3*params[0]+params[1]], '--', color=colorList[seqIx])
plt.legend(frameon=False)
plt.xlabel('Ligation % (Experimental)', fontsize=14)
plt.ylabel('Recovery Ratio (kSeq)', fontsize=14)

plt.show()
    
```

```python
def dropout_test(x,y,dropoutDepth=1):
    from itertools import combinations
    
    maskList = list(combinations([0,1,2,3,4], r=dropoutDepth))
    aList = []
    for itr in maskList:
        mask = [False if i in itr else True for i in range(5)]
        params = get_fitting_line(x[mask], y[mask])
        aList.append(params[0])
    return np.array(aList)

fig,ax = plt.subplots(1,3,figsize=[15,5])

for depth in [1,2,3]:
    dropoutSlope = []
    for seq in expData:
        dropoutSlope.append(dropout_test(seq['ligPct'], seq['kSeqRecRatio'],dropoutDepth=depth))
    bplot = ax[depth-1].boxplot(dropoutSlope, positions=[i for i in range(10)], patch_artist=True, )
    xtickNames = plt.setp(ax[depth-1], xticklabels=[seq['id'] for seq in expData])
    plt.setp(xtickNames, rotation=45)
    for boxIx,box in enumerate(bplot['boxes']):
        box.set(linewidth=0)
        box.set(facecolor = colorList[boxIx])
    for whisker in bplot['whiskers']:
        whisker.set(linewidth=1)
    for median in bplot['medians']:
        median.set(linewidth=1)
    for cap in bplot['caps']:
        cap.set(linewidth=1)
    for boxIx,flier in enumerate(bplot['fliers']):
        flier.set(markerfacecolor = colorList[boxIx], markersize=5, marker='o', markeredgecolor='none')
    ax[depth-1].set_xlabel('Dropout number = %i' %depth, fontsize=12)
ax[0].set_ylabel('Estimated slope', fontsize=12)
plt.tight_layout()
plt.show()
```

## Include error rate of k-seq estimated from drop-out test

```python
import numpy as np


def kA_fit(x,y):
    from scipy.optimize import curve_fit
    def func(x,k,A):
        return A*(1-np.exp(-k*x))
    
    try:
        params, pcov = curve_fit(func, xdata=x, ydata=y, method='trf')
    except RuntimeError:
        params = [np.nan, np.nan]
    return params

    
def dropout_test(x,y,dropoutDepth=1):
    from itertools import combinations
    
    maskList = list(combinations([0,1,2,3,4], r=dropoutDepth))
    paramList = [[], [], []]
    for itr in maskList:
        mask = [False if i in itr else True for i in range(5)]
        # print(x[mask], y[mask])
        params = kA_fit(x[mask], y[mask])
        paramList[0].append(params[0])
        paramList[1].append(params[1])
        paramList[2].append(params[0]*params[1])
    print(paramList[2])
    return np.array(paramList)

def dev_est():
    for seqIx,seq in enumerate(expData):
        print(seq['id'])
        paramList = dropout_test(x=np.array([1.5, 5, 16.5, 55, 180]), y=np.array(seq['kSeqRecRatio']), dropoutDepth=2)
        expData[seqIx]['dropoutStat'] = {
            'kMean': np.mean(paramList[0]),
            'kDev': np.std(paramList[0]),
            'AMean': np.mean(paramList[1]),
            'ADev': np.std(paramList[1]),
            'kAMean': np.mean(paramList[2]),
            'kADev': np.std(paramList[2])
        }

if __name__=='__main__':
    dev_est()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def kA_fit(x,y):
    from scipy.optimize import curve_fit
    def func(x,k,A):
        return A*(1-np.exp(-k*x))
    try:
        params, pcov = curve_fit(func, xdata=x, ydata=y, method='trf')
    except RuntimeError:
        params = [np.nan, np.nan]
    return params

    
def dropout_test(x,y,dropoutDepth=1):
    from itertools import combinations
    
    maskList = list(combinations([0,1,2,3,4], r=dropoutDepth))
    paramList = [[], [], []]
    for itr in maskList:
        mask = [False if i in itr else True for i in range(5)]
        # print(x[mask], y[mask])
        params = kA_fit(x[mask], y[mask])
        paramList[0].append(params[0])
        paramList[1].append(params[1])
        paramList[2].append(params[0]*params[1])
    # print(paramList[2])
    return np.array(paramList)                
```
