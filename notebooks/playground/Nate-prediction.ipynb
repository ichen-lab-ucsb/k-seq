## Pairwise distance

```python
import networkx as nx

def get_dist(seq1):
    
    def pairwise(seq):
        return Levenshtein.hamming(seq, seq1.name)
    
    return seq1.index.to_series().apply(pairwise)

A = pd.DataFrame(index=peak2_members, columns=peak2_members)
A = A.apply(get_dist, axis=0)
G = nx.from_pandas_adjacency(df=(A == 1))
```

```python
def load_seq_bootstrap_res(res_root, seq_to_hash, seq, stats):
    seq_res = read_json(f'{res_root}/{seq_to_hash[seq]}.json')
    try:
        return pd.read_json(seq_res['uncertainty']['record'])[stats].values
    except:
        return np.nan

def get_bootstrap_res(seq_list, stats='kA'):
    res_root = '/mnt/storage/projects/k-seq/working/byo-doped/least_square/pandaSeq/bootstrap_201912-202002/table-seq-in-all_bs-num-1000_bs-mtd-data_no-zero-false_inv-weight-false_core-40/results/seqs/'
    seq_to_hash = read_json(res_root + 'seq_to_hash.json')
    
    return pd.DataFrame({seq: load_seq_bootstrap_res(res_root, seq_to_hash, seq, stats) for seq in seq_list}).T
```

```python
seq_kA_res = get_bootstrap_res(seq_list=peak2_members, stats='kA')
```

```python
seq_kA_mean = byo_doped['kA_mean'][peak2_members]
```

```python
def normality_dAgostino_test(seq):
    return stats.normaltest(seq, nan_policy='propagate').pvalue

def normality_shapiro_test(seq):
    return stats.shapiro(seq)[0]

rawkA_normal_da = seq_kA_res.apply(normality_dAgostino_test, axis=1)
logkA_normal_da = np.log10(seq_kA_res).apply(normality_dAgostino_test, axis=1)
rawkA_normal_shapiro = seq_kA_res.apply(normality_shapiro_test, axis=1)
logkA_normal_shapiro = np.log10(seq_kA_res).apply(normality_shapiro_test, axis=1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))

pd.DataFrame({"d'Agostino-Pearson": [np.sum(rawkA_normal_da >= 0.05), np.sum(logkA_normal_da >= 0.05)],
 "Shapiro-Wilk": [np.sum(rawkA_normal_shapiro >= 0.05), np.sum(logkA_normal_shapiro >= 0.05)]
}, index=[r'$kA$', r'$\log_{10}(kA)$']).plot(kind='bar', width=0.5, ax=ax)
ax.set_ylabel('Seq failed to reject normal assumption\n(p ≥ 0.05)')
ax.set_title('Normality test', fontsize=16)
ax.legend(ncol=1, loc=(1, 0.8))
plt.show()
```

```python
from scipy import stats

def unpaired_welch_t_test(seq1, seq2, log=False):
    if log:
        return stats.ttest_ind(np.log10(seq_kA_res.loc[seq1]), np.log10(seq_kA_res.loc[seq2])).pvalue
    else:
        return stats.ttest_ind(seq_kA_res.loc[seq1], seq_kA_res.loc[seq2]).pvalue

def wilcoxon_rank_sum_test(seq1, seq2):
    return stats.ranksums(seq_kA_res.loc[seq1], seq_kA_res.loc[seq2]).pvalue

def mann_whitney_wilcoxon_test(seq1, seq2):
    return stats.mannwhitneyu(seq_kA_res.loc[seq1], seq_kA_res.loc[seq2], alternative='two-sided').pvalue
```

```python
welch_t_test_res = {
    edge: unpaired_welch_t_test(*edge) for edge in G.edges
}

welch_t_test_res_log = {
    edge: unpaired_welch_t_test(*edge, log=True) for edge in G.edges
}

wilcoxon_rank_sum_res = {
    edge: wilcoxon_rank_sum_test(*edge) for edge in G.edges
}

mann_whitney_wilcoxon_res = {
    edge: mann_whitney_wilcoxon_test(*edge) for edge in G.edges
}
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 4))

pd.DataFrame(
    {
        "Welch t-test": [
            np.sum(np.array(list(welch_t_test_res.values())) < 0.05),
            np.sum(np.array(list(welch_t_test_res.values())) >= 0.05)
        ],
        "Welch t-test (log)": [
            np.sum(np.array(list(welch_t_test_res_log.values())) < 0.05),
            np.sum(np.array(list(welch_t_test_res_log.values())) >= 0.05)
        ],
        "Wilcoxon rank sum": [
            np.sum(np.array(list(wilcoxon_rank_sum_res.values())) < 0.05),
            np.sum(np.array(list(wilcoxon_rank_sum_res.values())) >= 0.05)
        ],
        "Mann-Whitney-Wilcoxon": [
            np.sum(np.array(list(mann_whitney_wilcoxon_res.values())) < 0.05),
            np.sum(np.array(list(mann_whitney_wilcoxon_res.values())) >= 0.05)
        ]
    },
    index=['Significant', 'Non-significant']).plot(kind='bar', width=0.5, ax=ax)
ax.set_ylabel('Number of comparisons')
ax.set_title('Compare kA of one-dist neighbors \n(two-sided, p ≤ 0.05)', fontsize=16)

ax.legend(ncol=1, loc=(1, 0.8))
plt.show()
```
